```python
import pandas as pd
import tracemalloc

def measure_memory():
    tracemalloc.start()  
    
    
    chunks = pd.read_csv('D:/read_large_csv/2019-Oct.csv', chunksize=10000)
    for chunk in chunks:
        pass  

    peak_mem = tracemalloc.get_traced_memory()[1] / (1024 ** 2)
    tracemalloc.stop()  
    
    print(f"Peak memory used: {peak_mem:.2f} MB")

measure_memory()
```
